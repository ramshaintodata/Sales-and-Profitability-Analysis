# Statistical Summary:

The Sales data is highly skewed, with a few transactions reaching up to $11,199. Profit shows wide variation, ranging from a loss of $3,839 to a gain of $3,919, with a low average. Most orders include 2â€“5 items, though some reach up to 14. Discounts are typically 0% or 20%, but extreme discounts up to 80% impact profit. Postal Code and Row ID serve as identifiers and hold little analytical value.
The Superstore dataset includes three main product categories and 17 sub-categories, with Office Supplies, Binders, and Paper being the most common. Customers fall into three segments, with Consumers making up over 50% of sales. The West region leads in orders, with data from 310 cities across 46 states. California and New York City are top contributors. Standard Class is the dominant shipping method, used in over 50% of orders. Most IDs are unique, indicating a broad customer and product base.

**Outlier Detection:**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Superstore Dataset.csv", encoding='ISO-8859-1')
df["Profit Margin"] = df["Profit"] / df["Sales"].replace(0, 1e-5)
```

```python
numeric_features = ["Sales", "Profit", "Discount", "Quantity", "Profit Margin"]
```

```python
plt.figure(figsize=(15, 8))
for i, col in enumerate(numeric_features):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(data=df, y=col, color='skyblue')
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
plt.show()
```

```python
outlier_summary = {}
for col in numeric_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_summary[col] = len(outliers)
```

```python
print("Outlier Count by Feature:")
print(outlier_summary)
```

**CONDITIONAL ANALYSIS FOR SUPERSTORE DATA**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Superstore Dataset.csv", encoding='ISO-8859-1')
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df["Profit Margin"] = df["Profit"] / df["Sales"].replace(0, 1e-5)
df["High Discount"] = df["Discount"] > 0.3
df["Loss"] = df["Profit"] < 0
df["Delivery Delay"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
segment_region_pct = df.groupby(["Segment", "Region"]).size() / len(df) * 100
```

```python
avg_sales_profit = df.groupby(["Segment", "Region"])[["Sales", "Profit"]].mean()
```

```python
df["Discount Bin"] = pd.cut(df["Discount"], bins=[-0.01, 0.1, 0.2, 0.3, 0.5, 1.0])
discount_profit_cat = df.groupby(["Category", "Discount Bin"])["Profit"].mean().unstack()
```

```python
high_discount_loss_pct = df[df["High Discount"]]["Loss"].mean() * 100
```

```python
repeat_counts = df["Customer ID"].value_counts()
repeat_ids = repeat_counts[repeat_counts > 1].index
df["Repeat Buyer"] = df["Customer ID"].isin(repeat_ids)
repeat_spend = df.groupby("Repeat Buyer")[["Sales", "Profit"]].mean()
repeat_pct = df["Repeat Buyer"].mean() * 100
```

```python
loss_by_subcat = df[df["Loss"]].groupby("Sub-Category")[["Profit", "Sales"]].sum()
loss_by_subcat["Loss %"] = -loss_by_subcat["Profit"] / loss_by_subcat["Sales"] * 100
loss_by_subcat = loss_by_subcat.sort_values("Loss %", ascending=False)
```

```python
prob_profit = df.groupby("Discount Bin")["Profit"].apply(lambda x: (x > 0).mean())
```

```python
unprofitable_orders = df[df["Profit"] < 0]
unprofitable_cat_region = unprofitable_orders.groupby(["Category", "Region"]).size().unstack().fillna(0)
```

```python
avg_delay = df["Delivery Delay"].mean()
delay_by_seg_cat = df.groupby(["Segment", "Category"])["Delivery Delay"].mean()
```

```python

region_perf = df.groupby(["Segment", "Region"])[["Sales", "Profit"]].mean().unstack()
```

```python
subcat_stats = df.groupby("Sub-Category")[["Discount", "Profit"]].mean()
profitable_high_discount = subcat_stats[(subcat_stats["Discount"] > 0.3) & (subcat_stats["Profit"] > 0)]
```

```python
print(f" Repeat Buyer %: {repeat_pct:.2f}%")
```

```python
print(f" Average Delivery Delay: {avg_delay:.2f} days")
```

```python
print("\n % of Orders by Segment and Region:\n", segment_region_pct)
print("\n Avg Sales and Profit per Segment/Region:\n", avg_sales_profit)
print("\n Profitability across Discount Bins by Category:\n", discount_profit_cat)
print("\n Repeat Buyer Spend vs One-Time:\n", repeat_spend)
print("\n Sub-Categories with Highest % Loss:\n", loss_by_subcat[["Loss %"]])
print("\n P(Profit > 0) by Discount Bin:\n", prob_profit)
```

# After analyzing the Superstore data, a few key takeaways stand out. Discounts over 30% often lead to losses, especially in sub-categories like Tables and Bookcases, so itâ€™s worth rethinking our pricing strategy. The Consumer segment performs well across all regions, making it a good target for focused marketing. Repeat buyers spend more and make up a large part of revenue, so loyalty programs or special offers could help increase retention. Some sub-categories, like Binders and Paper, stay profitable even with discounts, these are great options for bundling or upselling. Overall, we need smarter discounting and better customer targeting to grow both sales and profit.
